# GridSearchCV Model

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
df.head()
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[[
    'koi_disposition', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration',
    'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 
    'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec',
    'koi_kepmag'
]]
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign X (data) and y (target)
X = selected_features.drop("koi_disposition", axis=1)
y = selected_features["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label-encode y data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
print(y)
```

```python
print(encoded_y)
```

```python
# Split the date between train and test values
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=1)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn import svm
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV

parameter_candidates = [
#     {'C': [1, 10], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001, 0.005, 0.0005], 'kernel': ['rbf']}
]
```

```python
# Create a classifier object with the classifier and parameter candidates
model_gridSearch = GridSearchCV(estimator=svm.SVC(), param_grid=parameter_candidates, n_jobs=-1)
```

```python
# Train the classifier on data1's feature and target data
model_gridSearch.fit(X_train, y_train) 
```

```python
# View the accuracy score
print('Train Best score:', model_gridSearch.best_score_)
```

```python
# View the best parameters for the model found using grid search
print('Best C:', model_gridSearch.best_estimator_.C) 
print('Best Kernel:', model_gridSearch.best_estimator_.kernel)
print('Best Gamma:', model_gridSearch.best_estimator_.gamma)
```

```python
# Make predictions with the hypertuned model
predictions = model_gridSearch.predict(X_test)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=['CANDIDATE', 'CONFIRMED', 'FALSE POSITIVE']))
```
