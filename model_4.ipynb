# Deep Learning Model

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
df.head()
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[[
    'koi_disposition', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration',
    'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 
    'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec',
    'koi_kepmag'
]]
selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign X (data) and y (target)
X = selected_features.drop("koi_disposition", axis=1)
y = selected_features["koi_disposition"]
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label-encode y data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
print(y)
```

```python
print(encoded_y)
```

```python
# Split the date between train and test values
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=1, stratify=encoded_y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the StandardScaler and perform some feature selection

```python
# Scale our data using StandardScaler
from sklearn.preprocessing import MinMaxScaler
from keras.utils import to_categorical

# Create a StandardScaler model and fit it to the training data
X_scaler = MinMaxScaler().fit(X_train)

# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
print(y_train_categorical)
print(y_test_categorical)
```

# Create Deep Learning Model



```python
from keras.models import Sequential
from keras.layers import Dense

# Create a Neural Network model here
model_DL = Sequential()
model_DL.add(Dense(units=100, activation='relu', input_dim=16))
model_DL.add(Dense(units=100, activation='relu'))
model_DL.add(Dense(units=3, activation='softmax'))
```

```python
# Compile and fit the model
model_DL.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model_DL.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=0
)
```

## Quantify our Trained Model

```python
model_DL_loss, model_DL_accuracy = model_DL.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Deep Learning Model - Loss: {model_DL_loss}, Accuracy: {model_DL_accuracy}")
```
